```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("home.csv")
```

```python
df
```

```python
plt.xlabel('Area in(sqr ft)')
plt.ylabel('Price in(US$)')
plt.scatter(df.area,df.price)
```

```python
#creating the obj of LinearRegression
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)# Fit() it means you training you LinearRegression Model
```

```python
reg.predict([[3300]])
```

```python
plt.xlabel('Area in(sqr ft)')
plt.ylabel('Price in(US$)')
plt.scatter(df.area,df.price)
plt.plot(df.area,reg.predict(df[['area']]))
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y=0
m=reg.coef_
x=3300
b=reg.intercept_
y=m*x+b
y
```
